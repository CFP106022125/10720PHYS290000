```python
#!/usr/bin/env python
# coding: utf-8

# In[4]:


import numpy as np
import matplotlib.pyplot as plt
import math

h=6.626*10**-34 # Planck constant
c=3*10**8       # speed of light
k=1.38*10**-23  # Boltzmann Constant JK^-1

v=np.linspace(10**0,10**15,100000,endpoint=True)
B=2*h*v**3/c**2/(math.e**(h*v/k/5000)-1)

#plt.errorbar(v,B,yerr=0.5*B ,xerr=0.1*v, ecolor"green")
num=100000                                                                                  
plt.plot(v,B)
#plt.xlim(10**15,10**15.2)
#plt.ylim(10**-14,10**-15)
#plt.plot(v,B)
plt.xlabel("Frequency (Hz)")
plt.ylabel("identity Watta/Hz/m$^2$")

plt.show()
dB=np.random.normal(0.,10**-8.5,num)
B+= dB

plt.plot(v,B)
plt.show()



```
