```python
import numpy as np
import matplotlib.pyplot as plt
import math
v=50*np.random.random(1)
theta=90/180*math.pi*np.random.random(1)
a=100*np.random.random(1)
b=50*np.random.random(1)
t=np.linspace(0,10,1000)
x=v*math.cos(theta)*t
y=v*math.sin(theta)*t-1/2*9.8*t**2
n=((x-a)**2+(y-b)**2)**(1/2)
if any(n-1<=0):
      print("打中了!")
else:
      print("沒打到...")
plt.plot(x,y,"b")
plt.plot(a,b,"ro")
plt.xlim(0,100)
plt.ylim(0,50)
plt.xlabel("distance(m)")
plt.ylabel("height(m)")
plt.show()
#print(n)

```

```python
import numpy as np
import matplotlib.pyplot as plt
import math
v=np.linspace(0,50,500)
theta=np.linspace(0,math.pi/2,500)
a=100*np.random.random(1)
b=50*np.random.random(1)
for k in range(1,500):
    for m in range(1,500):
        t=np.linspace(0,10,1000)
        x[v[k]][theta[m]]=v[k]*math.cos(theta[m])*t
y[k][m]=v[k-1]*math.sin(theta[m-1])*t-1/2*9.8*t**2
n=((x-a)**2+(y-b)**2)**(1/2)
if any(n-1<=0):
      print("打中了!")
else:
      print("沒打到...")
plt.plot(x,y,"b")
plt.plot(a,b,"ro")
plt.xlim(0,100)
plt.ylim(0,50)
plt.xlabel("distance(m)")
plt.ylabel("height(m)")
plt.show()
#print(n)
```
